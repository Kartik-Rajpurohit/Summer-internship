```python
import pandas as pd

dict1 = {"A": 10, "B": 50, "C": 80, "D": 100}
series = pd.Series(dict1)

squared_series = series.apply(lambda x: x ** 2)
added_series = series.apply(lambda x: x + 10)

ascending_series = series.sort_values(ascending=True)
descending_series = series.sort_values(ascending=False)
unique_values = series.unique()
n_unique_values = series.nunique()

print("Original Series:\n", series)
print("\nSquared Series:\n", squared_series)
print("\nAdded Series:\n", added_series)
print("\nAscending Series:\n", ascending_series)
print("\nDescending Series:\n", descending_series)
print("\nUnique Values:\n", unique_values)
print("\nNumber of Unique Values:\n", n_unique_values)
```

```python
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
pio.templates.default = "plotly_white"
```

```python
df.describe()
```

```python
fig = px.scatter(df, x='Price', y='Revenue generated', color='Product type', hover_data=['Number of products sold'], trendline='ols')
fig.show()
```

```python
sales_data = df.groupby('Product type')['Number of products sold'].sum().reset_index()
sales_data
```

```python
pie_chart = px.pie(sales_data, values= 'Number of products sold',
                   names='Product type',
                   title='Sales by Product type',
                   hover_data='Number of products sold',
                   hole=0.5,
                   color_discrete_sequence=px.colors.qualitative.Pastel)

pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

```python
total_revenue = df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
colors = ['#FF5733', '#33C1FF', '#28B463', '#F1C40F', '#9B59B6']
fig=go.Figure()
fig.add_trace(go.Bar(x=total_revenue['Shipping carriers'],
                     y=total_revenue['Revenue generated'],
                     marker_color=colors[:len(total_revenue)]
                     ))
fig.update_layout(title='Total revenue by Shipping Carrier',
                  xaxis_title='Shipping Carrier',
                  yaxis_title='Revenue Generated')
fig.show()
```

```python
from pickle import TRUE
avg_lead_time = df.groupby('Product type')['Lead time'].mean().reset_index()
avg_manufactuturing_costs = df.groupby('Product type')['Manufacturing costs'].mean().reset_index()
result = pd.merge(avg_lead_time, avg_manufactuturing_costs,
                  on="Product type")
result.rename(columns={'Lead time' : 'Average Lead Time',
                       'Manufacturing costs': 'Average Manufacturing Costs'},
              inplace=TRUE)
print(result)
```
